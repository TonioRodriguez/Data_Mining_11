### EDA o Exploratory Data Analysis

Proceso crítico e iterativo de investigar un conjunto de datos para descubrir sus principales características, patrones, anomalías y relaciones.

Se realiza principalmente a través de métodos de resumen estadístico y técnicas de visualización gráfica.

Acuñado por el matemático estadounidense John W. Tukey, el EDA no se enfoca en la formulación de modelos predictivos finales, sino en el entendimiento profundo de los datos.

En una analogía profesional, el EDA es equivalente a la labor de un detective que llega a la escena de un crimen: su primer objetivo no es arrestar a un sospechoso, sino examinar toda la evidencia disponible, identificar pistas, entender el contexto y formular teorías sobre lo que ocurrió.

Un EDA sistemático generalmente involucra las siguientes etapas:

  * Revisar las primeras y últimas filas del conjunto de datos.

  * Determinar las dimensiones - Patrones y variables (número de filas y columnas).

  * Cuantificar y visualizar la distribución de valores perdidos.

  * Identificar los tipos de datos de cada columna.

  * Utilizar resúmenes estadísticos (media, mediana, desviación estándar, cuantiles) para las variables numéricas.

  * Identificar y, si es necesario, tratar registros duplicados.

### 4. Outliers o valores atípicos

Un dato atípico u outlier es un punto de datos que difiere significativamente de otras observaciones. Es un valor que se encuentra a una distancia inusual del resto de los datos de un conjunto.

Método del Rango Intercuartílico (IQR)

Este método es robusto frente a otros outliers y se basa en los cuartiles de la distribución. Se utiliza frecuentemente para visualizar datos en un diagrama de caja (Box Plot) .

Un valor se considera un outlier si está:

  * Por debajo de: Q1 −1.5 × IQR

  * Por encima de: Q3 + 1.5 × IQR

Donde:

Q1 es el primer cuartil (percentil 25).

Q3 es el tercer cuartil (percentil 75).

IQR = Q3 − Q1.

```python
# Verify Outliers
import numpy as np
import pandas as pd

#attribute = 'sepal_length'
#attribute = 'sepal_width'
#attribute = 'petal_length'
attribute = 'petal_width'

# Quartiles
Q1 = data[attribute].quantile(0.25)
Q3 = data[attribute].quantile(0.75)
IQR = Q3 - Q1

# Limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Identifying outliers
outliers = data[(data[attribute] < lower_limit) | (data[attribute] > upper_limit)]
print(f"Outliers:{attribute}")
display(outliers)

num_outliers = len(outliers)
print(f"Total de Outliers encontrados: {num_outliers}")
```


### 8. Matriz de Correlación

  * Una matriz de correlación es una tabla que muestra los coeficientes de correlación entre diferentes variables (características) en un conjunto de datos.

  * Es una matriz que utiliza coeficientes de correlación (generalmente el Coeficiente de Correlación de Pearson, que mide la relación lineal). Estos coeficientes siempre caen en el rango de −1 a 1.

  * Proporciona un resumen conciso de las relaciones lineales que existen entre cada par de variables numéricas.

  * Es una matriz cuadrada donde las filas y columnas son las mismas variables. Los valores en la diagonal principal son siempre 1 (ya que una variable siempre se correlaciona perfectamente consigo misma).

  #### Interpretación del Coeficiente de Correlación (r)

  * r=1 (Correlación Positiva Perfecta):

    Interpretación: A medida que una variable aumenta, la otra aumenta exactamente en la misma proporción. Hay una relación lineal directa perfecta.

  * r=−1 (Correlación Negativa Perfecta):

    Interpretación: A medida que una variable aumenta, la otra disminuye exactamente en la misma proporción. Hay una relación lineal inversa perfecta.

  * r=0 (Sin Correlación Lineal):

    Interpretación: No hay relación lineal discernible entre las variables. Los cambios en una variable no se asocian linealmente con los cambios en la otra.

La matriz es crucial en  el análisis exploratorio:

  * Multicolinealidad (Correlación Alta): Si dos variables de entrada (X) tienen una correlación cercana a 1 o −1, son redundantes. Incluir ambas en modelos lineales (como la regresión lineal) puede causar inestabilidad. Se recomienda eliminar una de las variables para simplificar el modelo sin perder poder predictivo.

  * Relación con la Variable Objetivo (y): Si se incluye la variable objetivo (y) en la matriz, se puede identificar rápidamente qué variables de entrada (X) tienen la relación lineal más fuerte con la variable que se desea predecir.

Razón: El cálculo del coeficiente de Pearson implica la covarianza y la desviación estándar de las variables, operaciones que solo tienen sentido con valores numéricos.

**Correlación con Variables Categóricas**

Para medir la asociación entre variables categóricas o entre una numérica y una categórica, algunas opciones son:

  * Dos Variables Categóricas: Prueba de χ2(Chi-cuadrado) o el Coeficiente de Contingencia.

  * Una Numérica y una Categórica (con solo dos clases): Correlación Biserial-Puntual.

  * Una Numérica y una Categórica (con múltiples clases): ANOVA (Análisis de Varianza) para ver si la media de la variable numérica difiere significativamente entre las categorías.

### Gráficos de barras y tablas de contingencia (Atributos categóricos)


1. Gráfico de Barras (Bar Chart o Gráfico de Conteo)

  El gráfico de barras es la forma estándar de visualizar la distribución de una sola variable categórica (Análisis Univariado).

  * ¿Qué es? Un gráfico que presenta datos categóricos con barras rectangulares, donde la longitud o altura de la barra es proporcional a los valores que representa.

  * Representación:

    * Eje X: Muestra las categorías discretas de la variable (ej., "Hombre", "Mujer", "Clase 1", "Clase 2"). Las barras están separadas.

    * Eje Y: Muestra la frecuencia o el conteo de cuántas observaciones caen en cada categoría.

  * Interpretación: Permite ver rápidamente qué categorías son las más frecuentes (moda) y si existe un desbalance de clases.

  * Ejemplo: Un gráfico que muestra el número de pasajeros en el Titanic por cada clase de billete (Clase 1, Clase 2, Clase 3).


2. Tabla de Contingencia (Contingency Table o Tabla Cruzada)

La tabla de contingencia es una herramienta tabular (no visual) que se utiliza para examinar la relación o asociación entre dos o más variables categóricas (Análisis Bivariado o Multivariado).

  * ¿Qué es? Una disposición tabular de datos que resume la distribución de las frecuencias de las variables categóricas.

  * Representación:

    * Filas: Representan las categorías de la primera variable.

    * Columnas: Representan las categorías de la segunda variable.

    * Celdas: Contienen el conteo o la frecuencia conjunta (número de observaciones) de cada combinación de categorías.

  * Interpretación: Permite cuantificar la dependencia entre las variables.
  
  * Ejemplo, permite responder: "¿Cuántos hombres de la Clase 1 sobrevivieron?"

```python
contingency_table = pd.crosstab(data_categorical['sex'], data_categorical['embark_town'])
contingency_table
```

```python
# Visualize contingency_table
import seaborn as sns
import matplotlib.pyplot as plt

#Create a heatman
sns.heatmap(contingency_table, annot = True)
plt.show()
```

```python
contingency_table = pd.crosstab(data_categorical['sex'], data_categorical['embark_town'], margins = True)
contingency_table
```

```python
# Visualize contingency_table
import seaborn as sns
import matplotlib.pyplot as plt

#Create a heatman
sns.heatmap(contingency_table, annot = True)
plt.show()
```

```python
contingency_table_3 = data_categorical.groupby(['sex','class'],observed = True)['embark_town'].value_counts().unstack()
contingency_table_3
```

```python
# Visualize contingency_table
import seaborn as sns
import matplotlib.pyplot as plt

#Create a heatman
sns.heatmap(contingency_table_3, annot = True)
plt.show()
```
